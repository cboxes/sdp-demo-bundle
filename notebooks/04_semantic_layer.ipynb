# 📄 KPI Claims

```python
%sql
SHOW TABLES IN sdp_catalog.bronze;
```

```python
kpi_claims = spark.sql("""

SELECT
    city,
    SUM(claim_amount) as total_claims,
    AVG(resolution_days) as avg_resolution_days,
    SUM(fraud_flag) as suspicious_claims

FROM sdp_catalog.gold.fact_claims

GROUP BY city

""")
```

```python
kpi_claims.write.format("delta") \
    .option("overwriteSchema", "true") \
    .mode("overwrite") \
    .saveAsTable("sdp_catalog.semantic.kpi_claims")
```
